### Opennessを変化させた場合の精度の変化

```python
clean = [99.243, 98.18, 99.02]
noisy = [73.05, 58.11, 42.28] #各thresholdで最大値を使用
openness = ["0.087", "0.184", "0.293"]

plt.close()
plt.plot(clean, label="Erlangen Hospital Dataset(clean)")
plt.plot(noisy, label="Keio Hospital Dataset(noisy)")
plt.xticks([0, 1, 2], openness)
plt.legend()
plt.xlabel("Openness")
plt.ylabel("Accuracy [%]")
# plt.close()
```
